**Read the CSV File into a Pandas DataFrame**

Load the Stack Overflow survey data into a Pandas DataFrame:


```python
# Read the CSV file
df = pd.read_csv('survey-data.csv')

# Display the first few rows of the data
df.head()
```

**Create a SQLite Database and Insert the Data**

Now, let's create a new SQLite database (`survey-data.sqlite`) and insert the data from the DataFrame into a table using the sqlite3 library:


```python
import sqlite3

# Create a connection to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Write the dataframe to the SQLite database
df.to_sql('main', conn, if_exists='replace', index=False)


# Close the connection
conn.close()
```

**Verify the Data in the SQLite Database**
Verify that the data has been correctly inserted into the SQLite database by running a simple query:


```python
# Reconnect to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Run a simple query to check the data
QUERY = "SELECT * FROM main LIMIT 5"
df_check = pd.read_sql_query(QUERY, conn)

# Display the results
print(df_check)
```

## Demo: Running an SQL Query


Count the number of rows in the table named 'main'


```python
QUERY = """
SELECT COUNT(*) 
FROM main
"""
df = pd.read_sql_query(QUERY, conn)
df.head()
```

## Demo: Listing All Tables


To view the names of all tables in the database:


```python
QUERY = """
SELECT name as Table_Name FROM sqlite_master 
WHERE type = 'table'
"""
pd.read_sql_query(QUERY, conn)
```

## Demo: Running a Group By Query
    
For example, you can group data by a specific column, like Age, to get the count of respondents in each age group:


```python
QUERY = """
SELECT Age, COUNT(*) as count
FROM main
GROUP BY Age
ORDER BY Age
"""
pd.read_sql_query(QUERY, conn)
```

## Demo: Describing a table

Use this query to get the schema of a specific table, main in this case:


```python
table_name = 'main'

QUERY = """
SELECT sql FROM sqlite_master 
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY, conn)
print(df.iat[0,0])
```

## Hands-on Lab


### Visualizing the Distribution of Data

**Histograms**

Plot a histogram of CompTotal (Total Compensation).


```python
# --- SQL query: exclude NULL values in CompTotal ---
QUERY = """
SELECT * FROM main """

# --- Read the query into pandas ---
df = pd.read_sql_query(QUERY, conn)

# Exclude NA values which is count value greater than 30000
filtered_df = df["CompTotal"].dropna()
filtered_df = filtered_df[filtered_df <= 1000]

# Plot histogram
filtered_df.hist(
    bins=30,
    color="skyblue",
    edgecolor="black",
    figsize=(8, 5)
)
plt.title("Distribution of Total Compensation (CompTotal)")
plt.xlabel("Total Compensation")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
QUERY = """
SELECT CompTotal, Count(*) as Count FROM main 
GROUP BY CompTotal
ORDER BY CompTotal 
"""
pd.read_sql_query(QUERY, conn)
```

**Box Plots**

Plot a box plot of Age.


```python
QUERY = """
SELECT Age, COUNT(*) as count
FROM main
GROUP BY Age
ORDER BY Age
"""
print(pd.read_sql_query(QUERY, conn))

df = pd.read_sql_query(QUERY, conn)
ax = df.plot(kind="box", ylabel="Age Count", title="Age Box Plot")
ax.set_xticklabels(["All Respondents"])
```

### Visualizing Relationships in Data

**Scatter Plots**

Create a scatter plot of Age and WorkExp.


```python
%matplotlib inline
import seaborn as sns
```

```python
QUERY = """
SELECT * FROM main
"""
df = pd.read_sql_query(QUERY,conn)
plt.figure(figsize=(10, 5))
sns.scatterplot(x='Age', y='WorkExp', data=df)
plt.tight_layout()
plt.show()
```

**Bubble Plots**

Create a bubble plot of `TimeSearching` and `Frustration` using the Age column as the bubble size.


```python
QUERY = """
SELECT TimeSearching, Frustration, Age FROM main
"""
df=pd.read_sql_query(QUERY,conn)

sns.scatterplot(data=df, x='TimeSearching', y='Frustration', size='Age', hue='Age', alpha=0.5, sizes=(10, 500)) 

plt.title('TimeSearching and Frustation By Age', size=14) 
plt.xlabel('TimeSearching', size=10) 
plt.ylabel('Frustration', size=10)

plt.savefig("scatter_plot.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
QUERY = """
SELECT DISTINCT Frustration FROM main
"""
df=pd.read_sql_query(QUERY,conn)
pd.set_option('display.max_colwidth', None)
df.head(10)
```

### Visualizing Composition of Data

**Pie Charts**

Create a pie chart of the top 5 databases(`DatabaseWantToWorkWith`) that respondents wish to learn next year.


```python
QUERY = """
SELECT DatabaseWantToWorkWith,
    COUNT(*) AS COUNT
FROM main
WHERE DatabaseWantToWorkWith IS NOT NULL 
  AND TRIM(DatabaseWantToWorkWith) <> ''
GROUP BY DatabaseWantToWorkWith
ORDER BY COUNT DESC LIMIT 5
"""

df = pd.read_sql_query(QUERY, conn)
df.set_index('DatabaseWantToWorkWith', inplace=True) 

df['COUNT'].plot(kind='pie', figsize=(10,5), autopct='%1.1f%%', labels=None)  

plt.legend(labels=df.index, loc='upper right')
plt.title('Top 5 Databases Respondents Want To Work With') 
plt.axis('equal')
plt.show()
```

### Visualizing Comparison of Data

**Bar Chart**

Age Distribution


```python
QUERY = """
SELECT Age, COUNT(*) as count
FROM main
GROUP BY Age
ORDER BY Age
"""
print(pd.read_sql_query(QUERY, conn))

df = pd.read_sql_query(QUERY, conn)
plt.figure(figsize=(8,5))
plt.bar(df["Age"], df["count"], color="skyblue", edgecolor="black")

plt.title("Respondent Distribution by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Number of Respondents")

plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.savefig("age_chart.png", dpi=300, bbox_inches="tight")
plt.show()
```

Create a horizontal bar chart using the `MainBranch` column.


```python
QUERY = """
SELECT MainBranch, COUNT(*) as MainBranch
from main
group by MainBranch
"""

df=pd.read_sql_query(QUERY,conn)

df.plot(kind='barh', figsize=(10,6), color='lightskyblue')

plt.show()
```
